```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ga_avg = pd.read_csv("raw_soccer_files/player_ga_avg.csv")
passing = pd.read_csv("raw_soccer_files/player_passing_data.csv")
salaries = pd.read_csv("raw_soccer_files/player_salary_data.csv")
game_info = pd.read_csv("raw_soccer_files/game_info.csv")
team_info = pd.read_csv("raw_soccer_files/team_info.csv")
player_info = pd.read_csv("raw_soccer_files/player_info.csv")

print(f"ga_avg: {ga_avg.shape}")
print(ga_avg.count())
display(ga_avg.head())
print()

print(f"passing: {passing.shape}")
print(passing.count())
display(passing.head())
print()

print(f"salaries: {salaries.shape}")
print(salaries.count())
display(salaries.head())
print()

print(f"game_info: {game_info.shape}")
print(game_info.count())
display(game_info.head())
print()

print(f"team_info: {team_info.shape}")
print(team_info.count())
display(team_info.head())
print()

print(f"player_info: {player_info.shape}")
print(player_info.count())
display(player_info.head())
```

```python
ga_avg.describe()
```

```python
df = ga_avg.groupby(['player_id', 'season_name', 'team_id'])[['minutes_played', 'goals_added_above_avg_Dribbling',
       'goals_added_above_avg_Fouling', 'goals_added_above_avg_Interrupting',
       'goals_added_above_avg_Passing', 'goals_added_above_avg_Receiving',
       'goals_added_above_avg_Shooting', 'count_actions_Dribbling',
       'count_actions_Fouling', 'count_actions_Interrupting',
       'count_actions_Passing', 'count_actions_Receiving',
       'count_actions_Shooting', 'offensive_goals_added']].sum().reset_index()

df = pd.merge(df, player_info, on="player_id", how="left")
df = pd.merge(df, team_info, on="team_id", how="left")

df["goals_added_above_avg_Dribbling_90"] = 90 * df.goals_added_above_avg_Dribbling / df.minutes_played
df["goals_added_above_avg_Fouling_90"] = 90 * df.goals_added_above_avg_Fouling / df.minutes_played
df["goals_added_above_avg_Interrupting_90"] = 90 * df.goals_added_above_avg_Interrupting / df.minutes_played
df["goals_added_above_avg_Passing_90"] = 90 * df.goals_added_above_avg_Passing / df.minutes_played
df["goals_added_above_avg_Receiving_90"] = 90 * df.goals_added_above_avg_Receiving / df.minutes_played
df["goals_added_above_avg_Shooting_90"] = 90 * df.goals_added_above_avg_Shooting / df.minutes_played
df["offensive_goals_added_90"] = 90 * df.offensive_goals_added / df.minutes_played

print(df.shape)
df.head()
```

```python
passing["completed_passes"] = passing["attempted_passes"] * passing["pass_completion_percentage"]
passing["completed_passes"] = passing.completed_passes.apply(lambda x: round(x))
passing["xcompleted_passes"] = passing["attempted_passes"] * passing["xpass_completion_percentage"]

passing["total_distance_yds"] = passing["attempted_passes"] * passing["avg_distance_yds"]
passing["total_vertical_distance_yds"] = passing["attempted_passes"] * passing["avg_vertical_distance_yds"]

df2 = pd.merge(passing, player_info, on="player_id", how="left")
df2 = pd.merge(df2, team_info, on="team_id", how="left")
df2 = pd.merge(df2, game_info, on="game_id", how="left")
print(df2.shape)
df2.head()
```

```python
grouped = df2.groupby(["season_name", "team_id", "team_abbreviation", "player_id", 
                   "player_name", "general_position"]).game_id.count().reset_index()
grouped = grouped.rename(columns={"game_id": "game_count"})
pos = grouped.sort_values('game_count', ascending=False).groupby(["season_name", "team_id", "team_abbreviation", "player_id", "player_name"]).head(1)
pos = pos.rename(columns={"general_position": "primary_position"})
pos.head()
```

```python
# Aggregate Passing data

sub = df2.groupby(["season_name", "team_id", "team_abbreviation", "player_id", 
                   "player_name"]).agg({"attempted_passes": "sum", "completed_passes": "sum", "xcompleted_passes": "sum", 
                                                                            "passes_completed_over_expected": "sum", "total_distance_yds": "sum", "total_vertical_distance_yds": "sum",
                                                                           "share_team_touches": "mean"}).reset_index()
sub["passing_percentage"] = sub.completed_passes / sub.attempted_passes
sub["xpassing_percentage"] = sub.xcompleted_passes / sub.attempted_passes
sub["passes_completed_over_expected_p100"] = 100 * sub.passes_completed_over_expected / sub.attempted_passes 
sub["avg_distance_yds"] = sub.total_distance_yds / sub.attempted_passes
sub["avg_vertical_distance_yds"] = sub.total_vertical_distance_yds / sub.attempted_passes
sub = sub.fillna(0)

sub = pd.merge(sub, pos, on=["season_name", "team_id", "team_abbreviation", "player_id", 
                   "player_name"])
sub
```

```python
full = pd.merge(df, sub, on = ["season_name", "team_id", "team_abbreviation", "player_id", "player_name"], how="left")
full.info()
```

```python
full.loc[full.season_name == 2022].describe()
```

```python
full = full.loc[full.minutes_played > 0].reset_index(drop=True)
full.info()
```

```python
dfp = pd.DataFrame()

for year in range(2014, 2023):
    rolling = 1
    min_year = max(year-rolling, full.season_name.min())
    max_year = year - 1 
    
    yr_range = f"{min_year}-{max_year}"
    print(yr_range)
    
    x = full.loc[(full.season_name >= min_year) & (full.season_name <= max_year)]
    x = x.groupby(["player_id", "player_name", 'team_name', 'team_short_name', 'team_abbreviation', 'primary_position']).agg({"minutes_played": "sum", "attempted_passes": "sum", "completed_passes": "sum", "xcompleted_passes": "sum", 
                                                                            "passes_completed_over_expected": "sum", "total_distance_yds": "sum", "total_vertical_distance_yds": "sum",
                                                                           "share_team_touches": "mean", 'goals_added_above_avg_Dribbling_90': "sum",
       'goals_added_above_avg_Fouling_90': "sum", 'goals_added_above_avg_Interrupting_90': "sum",
       'goals_added_above_avg_Passing_90': "sum", 'goals_added_above_avg_Receiving_90': "sum",
       'goals_added_above_avg_Shooting_90': "sum", 'offensive_goals_added_90': "sum"}).reset_index()
    x["year_data"] = yr_range
    x["passing_percentage"] = x.completed_passes / x.attempted_passes
    x["xpassing_percentage"] = x.xcompleted_passes / x.attempted_passes
    x["passes_completed_over_expected_p100"] = 100 * x.passes_completed_over_expected / x.attempted_passes 
    x["avg_distance_yds"] = x.total_distance_yds / x.attempted_passes
    x["avg_vertical_distance_yds"] = x.total_vertical_distance_yds / x.attempted_passes
    x = x.fillna(0)
    
    y = full.loc[full.season_name == year][["season_name", "player_id", "player_name", 'goals_added_above_avg_Dribbling_90',
       'goals_added_above_avg_Fouling_90', 'goals_added_above_avg_Interrupting_90',
       'goals_added_above_avg_Passing_90', 'goals_added_above_avg_Receiving_90',
       'goals_added_above_avg_Shooting_90', 'offensive_goals_added_90']]
    
    new = pd.merge(x, y, on=["player_id", "player_name"], suffixes=["_pre", "_post"])
    
    dfp = pd.concat([dfp, new], ignore_index=True)
```

```python
cols = ['primary_position','share_team_touches',
       'passing_percentage', 'xpassing_percentage',
       'passes_completed_over_expected_p100', 'avg_distance_yds',
       'avg_vertical_distance_yds', 'goals_added_above_avg_Dribbling_90_pre',
       'goals_added_above_avg_Fouling_90_pre',
       'goals_added_above_avg_Interrupting_90_pre',
       'goals_added_above_avg_Passing_90_pre',
       'goals_added_above_avg_Receiving_90_pre',
       'goals_added_above_avg_Shooting_90_pre', 'offensive_goals_added_90_post']

features = pd.get_dummies(dfp.loc[:, cols])
target_cols = ['offensive_goals_added_90_post']
```

```python
# ML Helpers/Metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# scaling
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Algorithms
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.neighbors import KNeighborsRegressor

# trees
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import GradientBoostingRegressor

# fun ones
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
```

```python
# Create our train/test set
X = features.drop(columns=target_cols)
y = features["offensive_goals_added_90_post"]

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=42)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
def doMLRegression(model, X_train, y_train, X_test, y_test):
    # fit the model
    model.fit(X_train, y_train)

    # predict the model
    train_preds = model.predict(X_train)
    test_preds = model.predict(X_test)

    # get the TRAIN metrics
    resids = y_train - train_preds
    r2 = r2_score(y_train, train_preds)
    mse = mean_squared_error(y_train, train_preds)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_train, train_preds)

    # Print releveant metrics.
    print("TRAINING SET METRICS")
    print(f"The r2 is {r2}.")
    print(f"The mean squared error is {mse}.")
    print(f"The root mean squared error is {rmse}.")
    print(f"The mean absolute error is {mae}.")
    
    # get the TEST metrics
    resids = y_test - test_preds
    r2 = r2_score(y_test, test_preds)
    mse = mean_squared_error(y_test, test_preds)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, test_preds)

    # Print releveant metrics.
    print()
    print("TESTING SET METRICS")
    print(f"The r2 is {r2}.")
    print(f"The mean squared error is {mse}.")
    print(f"The root mean squared error is {rmse}.")
    print(f"The mean absolute error is {mae}.")

    # make a graph

    # Predicted vs Actual Graph
    plt.scatter(test_preds, y_test)
    plt.xlabel("Predictions")
    plt.ylabel("Observed")
    plt.title("Predicted vs Actual Plot")
    plt.plot(y_test, y_test)
    plt.show()


    # Residual Graph
    plt.scatter(test_preds, resids)
    plt.xlabel("Predictions")
    plt.ylabel("Residuals")
    plt.title("Residuals Plot")
    plt.plot(test_preds, np.zeros(len(test_preds)))
    plt.show()
```

```python
lr = LinearRegression()
doMLRegression(lr, X_train, y_train, X_test, y_test)
```

```python
en = ElasticNet()
doMLRegression(en, X_train, y_train, X_test, y_test)
```

```python
knn = KNeighborsRegressor(n_neighbors=15)
doMLRegression(knn, X_train, y_train, X_test, y_test)
```

```python
rf = RandomForestRegressor(random_state=42)
doMLRegression(rf, X_train, y_train, X_test, y_test)
```

```python
ada = AdaBoostRegressor(random_state=42)
doMLRegression(ada, X_train, y_train, X_test, y_test)
```

```python
xgb = XGBRegressor(random_state=42)
doMLRegression(xgb, X_train, y_train, X_test, y_test)
```

```python
lgb = LGBMRegressor(random_state=42)
doMLRegression(lgb, X_train, y_train, X_test, y_test)
```
