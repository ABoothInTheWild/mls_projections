## Imports

```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Data Preparation

```python
ga_avg = pd.read_csv("../raw_soccer_files/player_ga_avg.csv")
passing = pd.read_csv("../raw_soccer_files/player_passing_data.csv")
salaries = pd.read_csv("../raw_soccer_files/player_salary_data.csv")
game_info = pd.read_csv("../raw_soccer_files/game_info.csv")
team_info = pd.read_csv("../raw_soccer_files/team_info.csv")
player_info = pd.read_csv("../raw_soccer_files/player_info.csv")

print(f"ga_avg: {ga_avg.shape}")
print(ga_avg.count())
display(ga_avg.head())
print()

print(f"passing: {passing.shape}")
print(passing.count())
display(passing.head())
print()

print(f"salaries: {salaries.shape}")
print(salaries.count())
display(salaries.head())
print()

print(f"game_info: {game_info.shape}")
print(game_info.count())
display(game_info.head())
print()

print(f"team_info: {team_info.shape}")
print(team_info.count())
display(team_info.head())
print()

print(f"player_info: {player_info.shape}")
print(player_info.count())
display(player_info.head())
```

* Goals Added is uniquely identified by player, season, team and position
* Confirmed that players only play on one team per season
* Need to aggregate counting stats by player/season/team

```python
# Unique teams per player/season
ga_avg.groupby(["player_id", "season_name"]).team_id.nunique().sort_values(ascending=False)
```

```python
# Aggregate counting stats
df = ga_avg.groupby(['player_id', 'season_name', 'team_id'])[['minutes_played', 'goals_added_above_avg_Dribbling',
       'goals_added_above_avg_Fouling', 'goals_added_above_avg_Interrupting',
       'goals_added_above_avg_Passing', 'goals_added_above_avg_Receiving',
       'goals_added_above_avg_Shooting', 'count_actions_Dribbling',
       'count_actions_Fouling', 'count_actions_Interrupting',
       'count_actions_Passing', 'count_actions_Receiving',
       'count_actions_Shooting', 'offensive_goals_added']].sum().reset_index()

# add in metadata
df = pd.merge(df, player_info, on="player_id", how="left")
df = pd.merge(df, team_info, on="team_id", how="left")

# calculate per 90 stats
df["goals_added_above_avg_Dribbling_90"] = 90 * df.goals_added_above_avg_Dribbling / df.minutes_played
df["goals_added_above_avg_Fouling_90"] = 90 * df.goals_added_above_avg_Fouling / df.minutes_played
df["goals_added_above_avg_Interrupting_90"] = 90 * df.goals_added_above_avg_Interrupting / df.minutes_played
df["goals_added_above_avg_Passing_90"] = 90 * df.goals_added_above_avg_Passing / df.minutes_played
df["goals_added_above_avg_Receiving_90"] = 90 * df.goals_added_above_avg_Receiving / df.minutes_played
df["goals_added_above_avg_Shooting_90"] = 90 * df.goals_added_above_avg_Shooting / df.minutes_played
df["offensive_goals_added_90"] = 90 * df.offensive_goals_added / df.minutes_played

# calculate player season age
df["birth_date"] = pd.to_datetime(df.birth_date)
df["player_age"] = df.season_name - df.birth_date.dt.year

# calculate height in inches
df["height_in"] = df.height_ft * 12 + df.height_in
df.drop(columns=["height_ft", "birth_date"], inplace=True)

print(df.shape)
df.head()
```

## Game Logs Data Preparation

Interesting features here including passing, xpassing rates, as well as passing distances

Want to calculate aggregate stats based on weighted averages for these columns

```python
# Calculate counting stats
passing["completed_passes"] = passing["attempted_passes"] * passing["pass_completion_percentage"]
passing["completed_passes"] = passing.completed_passes.apply(lambda x: round(x))
passing["xcompleted_passes"] = passing["attempted_passes"] * passing["xpass_completion_percentage"]

passing["total_distance_yds"] = passing["attempted_passes"] * passing["avg_distance_yds"]
passing["total_vertical_distance_yds"] = passing["attempted_passes"] * passing["avg_vertical_distance_yds"]

# add in metadata
passing_full = pd.merge(passing, player_info, on="player_id", how="left")
passing_full = pd.merge(passing_full, team_info, on="team_id", how="left")
passing_full = pd.merge(passing_full, game_info, on="game_id", how="left")
print(passing_full.shape)
passing_full.head()
```

```python
# Aggregate Passing data
# "general_position": lambda x: x.mode()[0] gets the most used position by player
df2 = passing_full.groupby(["season_name", "team_id", "team_abbreviation", "player_id", 
                   "player_name"]).agg({"general_position": lambda x: x.mode()[0], "game_id": "nunique", "attempted_passes": "sum", "completed_passes": "sum", "xcompleted_passes": "sum", 
                                                                            "passes_completed_over_expected": "sum", "total_distance_yds": "sum", "total_vertical_distance_yds": "sum",
                                                                           "share_team_touches": "mean"}).reset_index()
df2.rename(columns={"general_position": "primary_position", "game_id": "games"}, inplace=True)

# Recalculate percentages from observed season stats
df2["passing_percentage"] = df2.completed_passes / df2.attempted_passes
df2["xpassing_percentage"] = df2.xcompleted_passes / df2.attempted_passes
df2["passes_completed_over_expected_p100"] = 100 * df2.passes_completed_over_expected / df2.attempted_passes 
df2["avg_distance_yds"] = df2.total_distance_yds / df2.attempted_passes
df2["avg_vertical_distance_yds"] = df2.total_vertical_distance_yds / df2.attempted_passes

df2 = df2.fillna(0)
df2
```

```python
salaries_sub = salaries.drop(["position", "mlspa_release", "competition"], axis=1).drop_duplicates()
df3 = salaries_sub.groupby(["player_id", "team_id", "season_name"]).agg({"base_salary": "max", "guaranteed_compensation": "max"}).reset_index()
df3.info()
```

## Merge OGA, Salary, & Game Datasets

```python
full = pd.merge(df, df2, on = ["season_name", "team_id", "team_abbreviation", "player_id", "player_name"], how="left")
full = pd.merge(full, df3, on = ["season_name", "team_id", "player_id"], how="left")

# reorder columns
cols = ['season_name', 'player_id', 'player_name', 'player_age',
       'height_in', 'weight_lb', 'nationality', 'base_salary', 'guaranteed_compensation', 'team_id', 'team_name', 'team_short_name',
       'team_abbreviation', 'primary_position', 'games', 'minutes_played', 'share_team_touches',
       'attempted_passes', 'completed_passes', 'passing_percentage', 
       'xcompleted_passes', 'xpassing_percentage', 'passes_completed_over_expected_p100',
       'passes_completed_over_expected', 'total_distance_yds', 'total_vertical_distance_yds', 
       'avg_distance_yds', 'avg_vertical_distance_yds', 'count_actions_Dribbling', 'count_actions_Fouling',
       'count_actions_Interrupting', 'count_actions_Passing',
       'count_actions_Receiving', 'count_actions_Shooting',
       'goals_added_above_avg_Dribbling', 'goals_added_above_avg_Fouling',
       'goals_added_above_avg_Interrupting', 'goals_added_above_avg_Passing',
       'goals_added_above_avg_Receiving', 'goals_added_above_avg_Shooting',
       'offensive_goals_added',  'goals_added_above_avg_Dribbling_90',
       'goals_added_above_avg_Fouling_90',
       'goals_added_above_avg_Interrupting_90',
       'goals_added_above_avg_Passing_90',
       'goals_added_above_avg_Receiving_90',
       'goals_added_above_avg_Shooting_90', 'offensive_goals_added_90'
      ]

full = full.loc[:, cols]
full.info()
```

## Null Imputation

```python
# this was caused by a divide by 0 - so just make the per 90 rate 0
full.loc[:, ['goals_added_above_avg_Dribbling_90', 'goals_added_above_avg_Fouling_90', 'goals_added_above_avg_Interrupting_90',
       'goals_added_above_avg_Passing_90', 'goals_added_above_avg_Receiving_90', 'goals_added_above_avg_Shooting_90', 'offensive_goals_added_90']] = full.loc[:, ['goals_added_above_avg_Dribbling_90',
       'goals_added_above_avg_Fouling_90', 'goals_added_above_avg_Interrupting_90', 'goals_added_above_avg_Passing_90',
       'goals_added_above_avg_Receiving_90', 'goals_added_above_avg_Shooting_90', 'offensive_goals_added_90']].fillna(0)
```

```python
# Use Average for player metadata
print(full.loc[:, ['player_age', 'height_in', 'weight_lb']].mean())
full["player_age"] = full["player_age"].fillna(26)
full["height_in"] = full["height_in"].fillna(70)
full["weight_lb"] = full["weight_lb"].fillna(165)
```

```python
# Use median for salary due to skew
print(full.loc[:, ['base_salary', 'guaranteed_compensation']].median())
full["base_salary"] = full["base_salary"].fillna(160000)
full["guaranteed_compensation"] = full["guaranteed_compensation"].fillna(179000)

# USA is most common nationality
print()
print(full.nationality.value_counts().head(1))
full["nationality"] = full["nationality"].fillna("USA")
```
