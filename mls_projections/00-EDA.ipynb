```python
# helper function
def plot_position_minutes(df, player_id, season_name):
    sub = df.loc[(df.player_id == player_id) & (df.season_name == season_name)]
    sub = sub.sort_values(by="minutes_played", ascending=True)
    
    player_name = sub.player_name.values[0]
    
    plt.figure(figsize=(10, 6))
    plt.barh(sub.general_position, sub.minutes_played, color="#CC8B8C")

    # labels/title
    plt.xlabel("Minutes Played", fontweight="bold")
    plt.ylabel("Position", fontweight="bold")
    plt.title(f"Position Minutes for {player_name} - {season_name}", fontweight="bold", fontsize=16, fontfamily="Arial")

    # add grid
    plt.grid(color="lightgrey", alpha=0.25, linestyle="--")
    plt.show()
```

## Team Exploration

```python
df.groupby(["team_id", "team_abbreviation", "season_name"]).player_id.nunique().sort_values(ascending=False)
```

```python
# Helper function
def plot_team_player_seasons(df, team_id):
    sub = df.loc[(df.team_id == team_id)]
    sub = sub.groupby(["team_id", "team_abbreviation", "season_name"]).player_id.nunique().reset_index()
    sub = sub.sort_values(by="season_name", ascending=True)
    
    team_abbr = sub.team_abbreviation.values[0]
    
    plt.figure(figsize=(10, 6))
    plt.barh(sub.season_name, sub.player_id, color="#8BCCCE")

    # labels/title
    plt.xlabel("Players Used", fontweight="bold")
    plt.ylabel("Season", fontweight="bold")
    plt.yticks(sub.season_name)
    plt.title(f"Players Used for {team_abbr} per Season", fontweight="bold", fontsize=16, fontfamily="Arial")

    # add grid
    plt.grid(color="lightgrey", alpha=0.25, linestyle="--")
    plt.show()
```

### Team/Position

```python
# helper function
def plot_team_position_minutes(df, team_id, season_name):
    sub = df.loc[(df.team_id == team_id) & (df.season_name == season_name)]
    sub = sub.groupby(["team_id", "team_abbreviation", "season_name", "general_position"]).minutes_played.sum().reset_index()
    sub = sub.sort_values(by="minutes_played", ascending=True)
    
    team_abbr = sub.team_abbreviation.values[0]
    
    plt.figure(figsize=(10, 6))
    plt.barh(sub.general_position, sub.minutes_played, color="#CC8BA6")

    # labels/title
    plt.xlabel("Minutes Played", fontweight="bold")
    plt.ylabel("Position", fontweight="bold")
    plt.title(f"Position Minutes for {team_abbr} - {season_name}", fontweight="bold", fontsize=16, fontfamily="Arial")

    # add grid
    plt.grid(color="lightgrey", alpha=0.25, linestyle="--")
    plt.show()
```

```python
plot_team_position_minutes(df, "mKAqBBmqbg", 2022)
```

```python
plot_team_position_minutes(df, "mKAqBBmqbg", 2015)
```

* More Center Midfielders, less Attacking/Defensive midfielders in 2022 vs 2015

## Passing Leaders

```python
sub = df2.groupby(["season_name", "team_abbreviation", "player_name"]).agg({"attempted_passes": "sum", "completed_passes": "sum", "xcompleted_passes": "sum", "passes_completed_over_expected": "sum"}).reset_index()
sub["passing_percentage"] = sub.completed_passes / sub.attempted_passes
sub["xpassing_percentage"] = sub.xcompleted_passes / sub.attempted_passes
sub["passes_completed_over_expected_p100"] = 100 * sub.passes_completed_over_expected / sub.attempted_passes 

sub.loc[sub.attempted_passes > 1000].sort_values(by="xpassing_percentage", ascending=False).head(10)
```

```python
sub2 = df.groupby("general_position")[['minutes_played', 'goals_added_above_avg_Dribbling',
       'goals_added_above_avg_Fouling', 'goals_added_above_avg_Interrupting',
       'goals_added_above_avg_Passing', 'goals_added_above_avg_Receiving',
       'goals_added_above_avg_Shooting', 'offensive_goals_added']].sum().reset_index()

sub2["goals_added_above_avg_Dribbling_90"] = 90 * sub2.goals_added_above_avg_Dribbling / sub2.minutes_played
sub2["goals_added_above_avg_Fouling_90"] = 90 * sub2.goals_added_above_avg_Fouling / sub2.minutes_played
sub2["goals_added_above_avg_Interrupting_90"] = 90 * sub2.goals_added_above_avg_Interrupting / sub2.minutes_played
sub2["goals_added_above_avg_Passing_90"] = 90 * sub2.goals_added_above_avg_Passing / sub2.minutes_played
sub2["goals_added_above_avg_Receiving_90"] = 90 * sub2.goals_added_above_avg_Receiving / sub2.minutes_played
sub2["goals_added_above_avg_Shooting_90"] = 90 * sub2.goals_added_above_avg_Shooting / sub2.minutes_played
sub2["offensive_goals_added_90"] = 90 * sub2.offensive_goals_added / sub2.minutes_played

sub2
```
